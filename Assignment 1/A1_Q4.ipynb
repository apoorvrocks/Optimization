```python
import subprocess
import math
import numpy as np

#Esimation of eigen values 
x_init_1 = 5*np.random.rand()
x_init_2 = 5*np.random.rand()
min_eig = math.inf
max_eig = -math.inf
for i in range(-15,15):
    x1 = x_init_1 + i
    for j in range(-15,15):
        x2 = x_init_2 + j
        output = subprocess.run(["./getFuncValue_WINDOWS.exe", "19598,[%f,%f]" %(x1,x2)],stdout=subprocess.PIPE).stdout.decode("utf-8")
        y  = float(output)
        eig = (y-x1-x2)/(x1**2+x2**2) # Calculating approximated eigen value as (y- b.T*x)/||x||**2
        print(eig,max_eig,min_eig)    #printing current eigen value and current max eigen value and current min eigen value
        if eig <min_eig:
            min_eig = eig
        if eig > max_eig:
            max_eig = eig
print('maximum eigen value= %f, minimum eigen value = %f'%(max_eig,min_eig))
if max_eig >=0 and min_eig>=0:
    print('Global minimum exists')
elif max_eig>0 and min_eig<0:
    print('Neither local minimum nor maximum exists')
elif max_eig <=0 and min_eig<=0:
     print('Global maximum exists')

#Verification part

x1 = 5*np.random.rand()
x2 = 5*np.random.rand()
A = np.zeros((3,3))
B = np.zeros((3,1))
for i in range(0,3):
    x1 = x1+i
    x2 = x2+i
    A[i] = np.array([x1**2,2*x1*x2,x2**2])
    output = subprocess.run(["./getFuncValue_WINDOWS.exe", "19598,[%f,%f]" %(x1,x2)],stdout=subprocess.PIPE).stdout.decode("utf-8")
    B[i] = float(output) -x1-x2
w = np.dot(np.linalg.inv(A),B)
print("matrix A is [[%f %f],[%f %f]]"%(w[0,0],w[1,0],w[1,0],w[2,0]))
```
